利用 Request + BeatifulSoup 爬取下列兩個網站內容並解析：

### 1. Dcard 網址： https://www.dcard.tw/f

```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://www.dcard.tw/f'
r = requests.get(url)
print(type(r))

# response 編碼方式 
r.encoding = 'utf-8'
print(r.text[0:3000])
```

```python
print('Request 取回之後該怎麼取出資料，資料型態是什麼？ =>')

html = r.text
print(type(html))
# BeautifulSoup 的第一個參數必須是 字串(因此 response 要先加 text 變成 response.text 再傳進去)
# Reference: https://tinyurl.com/y5fwpfnu
soup = BeautifulSoup(html, 'html.parser')
print(soup)
```

```python
print('為什麼要使用 BeatifulSoup 處理？處理後的型態是什麼？ => ')
# 使用 BeautifulSoup 剖析文字
soup = BeautifulSoup(html, 'html.parser')
print("BeautifulSoup 格式: %s"%type(soup))
```

### 2. 知乎： https://www.zhihu.com/explore

```python
url = 'https://www.zhihu.com/explore'
r = requests.get(url)
r.encoding = 'utf-8'

print(r.text[0:600])
```

### 3. 觀察一下知乎回來的資料好像有點怪怪的，該怎麼解決？

```python
import requests
url = 'https://www.zhihu.com/explore'

# headers 是一個 dict。user-agent 中間有個 Dash。user-agent value 可以隨意填寫
headers = {"user-agent": "OMG/0.0.1"}
r = requests.get(url, headers=headers)

r.encoding = 'utf-8'
print(r.text[0:600])
print("觀察一下知乎回來的資料好像有點怪怪的，該怎麼解決？")
print("解法：加 headers 參數")
```
