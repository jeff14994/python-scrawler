## Grab + PyQuery

```python
# 將 Day9 的程式改寫成 function
# 將圖片轉換 URL 格式
def get_imgur_img_url(img_tag):
    # 組合圖片而非網站的網址
    img_id = img_tag.split('/')[-1]
    # print(img_id)
    img_url = "https://i.imgur.com/%s"%(img_id + '.jpg')
    # print(img_url)
    return img_url
# 下載圖片
def download_img(img_url):
    output_dir = 'download'
    # check if download exists
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)
    # 對圖片送出請求
    with requests.get(img_url, stream=True) as r:
        # returns an HTTPError object if an error has occurred during the process.
        r.raise_for_status()
        # 檢查圖片
        img = Image.open(r.raw)
        # 檔名儲存（檔案格式可以大寫、也可以小寫）
        img_savename = "%s/%s.%s"%(output_dir, img_url.split('/')[-1], img.format)
        # 檔案儲存
        img.save(img_savename)
        print('Save image: %s'%img_savename)
```

```python
from grab import Grab
from pyquery import PyQuery as pq
#  得到 Response
g = Grab()
# print(g)
# Set Cookies (https://tinyurl.com/y4fx2f6e)
g.cookies.set(name='over18', value='1', domain='www.ptt.cc')
resp = g.go("https://www.ptt.cc/bbs/Beauty/M.1556291059.A.75A.html")
# print(resp.body)

# Parse the response 
doc = pq(resp.body)
title = doc('title')
body = doc('body')
# print(title.text())
# print(body.text())
links_all = doc("a")
# print(links)
data = {}
# print(type(links_all.items())) # type: <class 'generator'>
for a in links_all.items():
    # a.text().strip()
    # Python String strip() function will remove leading and trailing whitespaces
    title = a.text().strip()
    if 'imgur' not in title:
        continue
    # a.attr['href'] is the value of the href
    data[title] = a.attr['href']
    # get imgur url
    img_url = get_imgur_img_url(title)
    # download image 
    download_img(img_url)
```
