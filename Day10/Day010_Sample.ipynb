# 其他工具：Grab、PyQuery


* 利用 Grab 套件的存取 HTML 資源
* 利用 PyQuery 套件的解析 HTML 格式


## Grab

```python
# Download command: pip install -U grab
# URL: https://tinyurl.com/pfcgpzm
from grab import Grab
g = Grab()
resp = g.go('https://google.com')
resp.body
```

## PyQuery

```python
# Download command
# conda install -c anaconda pyquery
from pyquery import PyQuery as pq
doc = pq(resp.body)
title = doc('title')
print(type(title), title.text())
print(title.text())
```
