## 1. 請問高雄市有多少地區有溫度資料？

```python
import xmltodict 
with open('./data/64_72hr_CH.xml') as xml:
    data = dict(xmltodict.parse(xml.read()))
locations = data['cwbopendata']['dataset']['locations']['location']
total = 0
for location in locations:
#     顯示所有縣市
#     print(region['locationName'])
#     取得該地區天氣
    weathers = location['weatherElement']
    for weather in weathers:
#         判斷是否有溫度量測
        if(weather['description'] == "溫度"):
            total += 1
print("%d 個地區"%total)
```

## 2. 請取出每一個地區所記錄的第一個時間點跟溫度
