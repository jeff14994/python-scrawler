## 作業目標

### ① 取出今天所有的發文

```python
import requests
from bs4 import BeautifulSoup
# get url soup(Parsed)
url = "https://www.ettoday.net/news/news-list.htm"
def get_soup(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    return soup

soup = get_soup(url)
total = 0
# Find specific target
for data in soup.find(class_="part_list_2").find_all("h3"):
    date = data.find(class_="date").text
    content = data.find_all('a')[-1].text
    print(date, content)
    total += 1
print("共100則新聞: %d"%total)
```

```python
# 用 list 儲存(?) -> 用 dict 儲存更好(因為是鍵值關係)
# 先將所有分類列印出來，看有哪些類別
soup = get_soup(url)
catagories = []
for data in soup.find(class_="part_list_2").find_all('h3'):
    catagories.append(data.find(class_="tag").text)
# print(catagories)
all_catagories = set(catagories)
# print(all_catagories)
content_dict = {}
for data in soup.find(class_="part_list_2").find_all('h3'):
    catagory = data.find(class_="tag").text
    text = data.find_all("a")[-1].text
    # 使用 dict 儲存資料。如果剛開始分類不在 dict 裡先創一個分類 key 並宣告他的值為 list ，再將資料儲存進去
    if catagory not in content_dict:
        content_dict[catagory] = list()
        content_dict[catagory].append(text)
    else: 
        content_dict[catagory].append(text)
# print(content_dict.keys())
# print(content_dict)

# 得到每一個 dict value 裡的每一個 list 的長度(使用 map)
# map 第一個參數是 function ，第二個參數是 input variables ，依序餵給 function 處理
count = sum(map(len, content_dict.values()))
print(count)
```

### ③ 哪一個類別的文章最多

```python
# print(content_dict.keys())
# 找出哪一個類別的文章最多
def most_articles(content_dict):
    key_largest = 0
    for key in content_dict.keys():
        count = len(content_dict[key])
        if (count > key_largest):
            key_largest = count
            key = key
    return key, key_largest
# 印出最多篇
winner = most_articles(content_dict)
print("最多篇文章為: %s，共%d篇"%(winner[0], winner[1]))
print("\n")
# 總共文章數
count_all = sum(map(len, content_dict.values()))
print("本頁總共有 %d 篇文章"%count_all)
print("\n")
# 每一篇文章共有幾篇：
for key in content_dict.keys():
    count = len(content_dict[key])
    print("%s: 共%d篇"%(key, count))
```
