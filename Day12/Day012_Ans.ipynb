### ① 取出今天所有的發文

```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)

soup = BeautifulSoup(r.text, "html.parser")

for d in soup.find(class_="part_list_2").find_all('h3'):
    print(d.find(class_="date").text, d.find_all('a')[-1].text)
```

### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python
url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)

soup = BeautifulSoup(r.text, "html.parser")

# res 是一個 dict
res = {}

for d in soup.find(class_="part_list_2").find_all('h3'):
    date = d.find(class_="date").text
    title = d.find_all('a')[-1].text
#     tag = d.find(class_='tag').text
    # select()就跟find_all()是一樣的，只是內容的寫法不同，find_all()在上一年文章已經說明過了，這裡就不贅述了
    # https://tinyurl.com/ya8d4d7y
    # 撈出所有 class 是 tag 的內容
    tag = d.select('.tag')[0].text
    # setdefault 返回可以使用的 key 所對應的 value，若 key 不可用，提供默認值
    # e.g. 若 tag 可用，返回所對應的 value；若tag 不可用，返回 []
    res.setdefault(tag, [])
    # 把東西 append 到該 key 上
    res[tag].append({
        'title': title,
        'date': date
    })
    
print(res)
```

### ③ 哪一個類別的文章最多

```python
# ③ 哪一個類別的文章最多
data = []
for i in res:
#     print(i, len(res[i]))
    d = {}
    d['tag'] = i
    d['count'] = len(res[i])
    data.append(d)
# list 裡面存 dict ，每個 dict 的 key 分別為 tag 與 count
print(data)
# 
# [::-1] 表 reverse 整個 list
# sorted 表排序（由小到大）
# 取 key 為 count 的來排序，key 使用 lambda 將每一元素取 count 
#  Resources: https://tinyurl.com/y7yf462l
for d in sorted(data, key=lambda d: d['count'])[::-1]:
    print(d)
```

### list 取值與逆向

```python
# [::-1]順序相反操作
# [-1]讀取倒數第一個元素
# [3::-1]從下標為3（從0開始）的元素開始翻轉讀取
# Resources: https://tinyurl.com/y8474o9x
a = [1, 2, 3, 4, 5]
b = a[::-1]
c = a[-1]
d = a[3::-1]
print(a)
print(b)
print(c)
print(d)
```
