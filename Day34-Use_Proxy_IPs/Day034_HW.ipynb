# 反爬：代理伺服器/IP

* 「IP 黑/白名單」反爬蟲的因應策略

## 作業目標

* 目前程式中的 proxy_ips 是手動輸入的，請根據 https://free-proxy-list.net/ 寫一個可自動化抓取可用 Proxy 的 proxy_ips。




```python
import requests
from bs4 import BeautifulSoup
import random
```

```python
def get_proxy_ip(url):
    proxy_ips = []
    # 發起請求
    response = requests.get(url)
    # 剖析
    soup = BeautifulSoup(response.text, "html.parser")
    # print(soup)
    # 使用 id 去取 table
    table = soup.find(id="proxylisttable")
    # 將該table 所有 td tag 取出來
    all_proxy = table.findChildren('tr',recursive=True)
    # print(all_proxy)
    # 將所有 IP 變成 list
    all_ip = []
    for i in range(len(all_proxy)):
        # 去除第一個欄位
        if all_proxy[i].td == None:
            continue
        # 將所有 td tag 取出來 [:2] 表示取 td tag list 的前兩個元素（即 IP 與 port）
        all_list = all_proxy[i].findChildren('td',recursive=True)[:2] 
        # print(all_list)
        # 將個別 ip:port 綁在一起成字串
        ip = ""
        for i in range(len(all_list)):
            ip += str(all_list[i].text)
            # 中間加 ":"
            if(i == 0):
                ip += ":"
        all_ip.append(ip)
    # print(all_ip)
    return all_ip

    
```

```python
url = "https://free-proxy-list.net/"
proxy_ips = get_proxy_ip(url)
proxy_ips
```

```python
# 使用 IP
for i in range(10):
    ip = random.choice(proxy_ips)
    print('Use', ip)
    try:
        # timeout: You can tell Requests to stop waiting for a response after a given number of seconds with the timeout parameter. 
            # Resources: https://requests.readthedocs.io/en/master/user/quickstart/
        # proxies: If you need to use a proxy, you can configure individual requests with the proxies argument to any request method:
        # 使用 http 或 https
            # Resources: https://requests.readthedocs.io/en/master/user/advanced/
        resp = requests.get('http://ip.filefab.com/index.php',
                        proxies={'http': ip, 'https': ip}, timeout=10)
        soup = BeautifulSoup(resp.text, 'html.parser')
        print("Success!")
        print(soup.find('h1', id='ipd').text.strip())
        print("\n")
    except:
        print('Fail\n')
```
