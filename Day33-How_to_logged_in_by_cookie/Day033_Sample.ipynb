## 未授權前存取

```python
import requests
from bs4 import BeautifulSoup
```

```python
res = requests.get('https://www.ptt.cc/bbs/Gossiping/index.html', verify = False)
soup = BeautifulSoup(res.text, 'html.parser')
print(soup.text)
```

## 模仿授權機制

### 方法一：先送登入，再送請求

```python
# The Session object allows you to persist certain parameters across requests.
rs = requests.session()
payload={
    'from':'/bbs/Gossiping/index.html',
    'yes': 'yes'
}
res = rs.post('https://www.ptt.cc/ask/over18',verify = False, data = payload)
res = rs.get('https://www.ptt.cc/bbs/Gossiping/index.html',verify = False)
soup = BeautifulSoup(res.text,'html.parser')
print(soup.text) 
```

```python
# The Session object allows you to persist certain parameters across requests.
# Resources: https://requests.readthedocs.io/en/master/user/advanced/
s = requests.Session()
# s = requests
s.get('https://httpbin.org/cookies/set/sessioncookie/123456789')
r = s.get('https://httpbin.org/cookies')

print(r.text)
# '{"cookies": {"sessioncookie": "123456789"}}'
```

### 方法二：在請求中帶入 cookie

```python
# Requests verifies SSL certificates for HTTPS requests, just like a web browser.
# Resources: https://requests.readthedocs.io/en/master/user/advanced/
cookies = {'over18': '1'}
res = requests.get('https://www.ptt.cc/bbs/Gossiping/index.html',verify = False, cookies=cookies)
soup = BeautifulSoup(res.text,'html.parser')
print(soup.text)
```
