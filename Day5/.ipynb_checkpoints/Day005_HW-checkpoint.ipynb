# API 資料串接 - 以 Dcard API 實作範例

* 了解 Dcard API 使用方式與回傳內容
* 撰寫程式存取 API 且解析 JSON 格式資料


## 作業目標

* 請利用 API: https://www.dcard.tw/_api/forums/pet/posts?popular=true 回答下列問題：

1. 這個 API 一次會回傳幾筆資料？每一筆資料包含哪些欄位？
2. 取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」
3. 計算熱門/非熱門文章的「平均留言人數」與「平均按讚人數」

### 1. 這個 API 一次會回傳幾筆資料？每一筆資料包含哪些欄位？

```python
import requests
import json
response = requests.get("https://www.dcard.tw/_api/forums/job/posts?popular=true")
# Raw Data
response_content = response.text
# 前處理將 raw data 變成 json 檔比較容易看的懂(https://tinyurl.com/y3369qy8)
response_json = response.json()
print("總共回傳: %d 筆資料"%len(response_content))
print("\n")
print("欄位: ")
response_json
```

### 2. 取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」

```python
import requests
import json
response = requests.get("https://www.dcard.tw/_api/forums/job/posts?popular=true")
response_content = response.text
response_json = json.loads(response_content)
for data in response_json:
    print("標題:    %s"%data['title']) 
    print("發文時間: %s"%data['createdAt'])
    print("評論個數: %s"%data['commentCount'])
    print("按讚個數: %s"%data['likeCount'])
    print("\n")
```

### 3. 計算熱門/非熱門文章的「平均留言人數」與「平均按讚人數」

```python
import requests
import json
response = requests.get("https://www.dcard.tw/_api/forums/job/posts?popular=true")
response_json = json.loads(response.text)
amount = 0
commentCountTotal = 0
likeCountTotal = 0
for data in response_json: 
    amount +=1
    commentCountTotal += data['commentCount']
    likeCountTotal += data['likeCount']
#     print("評論個數: %s"%data['commentCount'])
#     print("按讚個數: %s"%data['likeCount'])
print("熱門平均留言人數：%d"%(commentCountTotal/amount))
print("熱門平均留按讚人數：%d"%(likeCountTotal/amount))

    
```

```python
# 寫成函式
import requests
import json
# 輸入網址獲取 json Data
def get_json_data_from_url(url):
    response = requests.get(url)
    response_json = json.loads(response.text)
    return response_json
# 印出熱門、按讚人數平均
def return_average(response_json):
    amount = 0
    commentCountTotal = 0
    likeCountTotal = 0
    for data in response_json: 
        amount +=1
        commentCountTotal += data['commentCount']
        likeCountTotal += data['likeCount']
    print("熱門平均留言人數：%d"%(commentCountTotal/amount))
    print("熱門平均留按讚人數：%d"%(likeCountTotal/amount))

# main function
if __name__ =='__main__': # Resources: https://tinyurl.com/ybbjp3rt
    # 印出熱門文章的「平均留言人數」與「平均按讚人數」
    print("熱門文章的「平均留言人數」與「平均按讚人數」")
    url = "https://www.dcard.tw/_api/forums/job/posts?popular=true"
    return_average(get_json_data_from_url(url))
    print("\n")
    # 印出非熱門文章的「平均留言人數」與「平均按讚人數」
    print("非熱門文章的「平均留言人數」與「平均按讚人數」")
    url = "https://www.dcard.tw/_api/forums/job/posts?popular=false"
    return_average(get_json_data_from_url(url))
```
