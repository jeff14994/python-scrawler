## 1. 取出今天所有的新聞(今天是 2020.12.23 我打算取 2020.12.22 [前一天，因為新聞篇數已固定])

- 透過調整 control_chrome 函式裡的 for 迴圈與 window.scrollTo() 參數
- 我發現，for-loop 70 次、window.scrollTo 第二個參數設 70000 可以找到 2020.12.22 整天的文章

```python
#  匯入工具
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
import time
#  使用 Chromedrive 開起瀏覽器
def open_chrome(url):
    browser  = webdriver.Chrome(executable_path="../chromedriver/chromedriver")
    browser.get(url)
    return browser
# 控制瀏覽器
def control_chrome(browser):
    for i in range(70):
        time.sleep(0.5)
        # window.scrollTo resource: https://tinyurl.com/y8vm7c26
        browser.execute_script("window.scrollTo(0, 70000);")
#  開始爬蟲
def crawl(browser):
#     total = 0
    html_source = browser.page_source
    soup = BeautifulSoup(html_source, "html.parser")
#     for data in soup.find(class_="part_list_2").find_all("h3"):
#         total += 1
#         print(data.find(class_="date").text, data.find_all('a')[-1].text)
#     print(total)
    return html_source
```

```python
# 取 2020.12.22 的新聞
url = "https://www.ettoday.net/news/news-list-2020-12-22-0.htm"
browser = open_chrome(url)
control_chrome(browser)
html_source = crawl(browser)
browser.close();
```

## 2. 取出現在時間兩小時內的新聞(我取 2020.12.22 1800-2000)

```python
url = "https://www.ettoday.net/news/news-list-2020-12-22-0.htm"
# 開瀏覽器
browser = open_chrome(url)
# 開始瀏覽
control_chrome(browser)

# 爬蟲
html_source = crawl(browser)

# 設定過濾條件(1800-2000)
soup = BeautifulSoup(html_source, "html.parser")
total = 0
for data in soup.find(class_="part_list_2").find_all("h3"):
#     print(data.find(class_="date").text, data.find_all('a')[-1].text)
    hour = data.find(class_="date").text[-6:-3]
    minute = data.find(class_="date").text[-2:]
#     print(minute)
#     print(eval(hour))
    try:
        if ( 18 <= eval(hour) <= 20 ):
            if (eval(hour) == 20):
                if(eval(minute) == 0):
                    print(data.find(class_="date").text, data.find_all('a')[-1].text)
                    total += 1
            else:
                print(data.find(class_="date").text, data.find_all('a')[-1].text)
                total += 1
                
    except:
        continue
print("===================================")        
print("2020.12.22 1800-2000 總共有 %d 篇新聞"%total) 

# 關閉瀏覽器
browser.close();
```

## 3. 根據範例，取出三天前下午三點到五點的新聞

```python
# 取 2020.12.19 1500-1700 的新聞
url = "https://www.ettoday.net/news/news-list-2020-12-19-0.htm"
# 開瀏覽器
# browser = open_chrome(url)
# 開始瀏覽
# control_chrome(browser)

# 爬蟲
# html_source = crawl(browser)

# 設定過濾條件(1800-2000)
soup = BeautifulSoup(html_source, "html.parser")
total = 0
for data in soup.find(class_="part_list_2").find_all("h3"):
#     print(data.find(class_="date").text, data.find_all('a')[-1].text)
    date = data.find(class_="date").text[-8:-6]
#     print(date)
    hour = data.find(class_="date").text[-6:-3]
    minute = data.find(class_="date").text[-2:]
#     print(minute)
#     print(eval(hour))
    # 指定日期
    desire_date = 19
    # 指定開始時間
    start_time = 15
    # 指定結束時間
    end_time = 17
    try:
        # 先判斷日期
        if (eval(date) == desire_date):
            # 再判斷起、迄時間
            if ( start_time <= eval(hour) <= end_time ):
                if (eval(hour) == end_time):
                    # 最後判斷 end_time 0 分的情況
                    if(eval(minute) == 0):
                        print(data.find(class_="date").text, data.find_all('a')[-1].text)
                        total += 1
                else:
                    print(data.find(class_="date").text, data.find_all('a')[-1].text)
                    total += 1
                
    except:
        continue
print("===================================")        
print("爬蟲結果：\n2020.12.%s %s00-%s00 總共有 %d 篇新聞"%(desire_date, start_time, end_time, total))

# 關閉瀏覽器
# browser.close();
```
